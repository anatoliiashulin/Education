## Нейросеть

```python
import matplotlib.pyplot as plt
```

```python
x_train = torch.rand(1000)
x_train = x_train * 10.0 -10.0


y_train = torch.cos(x_train)**2

plt.figure(figsize=(10,5))
plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('$ y = cos^2(x) $')
plt.show()
```

```python
noisy = torch.rand(y_train.shape) / 15.

plt.figure(figsize = (10,5))
plt.plot(x_train.numpy(), noisy.numpy(), 'o')
plt.title('Шумы')
plt.show()
```

```python
y_train = y_train+noisy
plt.figure(figsize=(10,5))
plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('$ y = cos^2(x) $')
plt.show()
```

```python
x_train.unsqueeze_(1)
y_train.unsqueeze_(1)
```

```python
x_val = torch.linspace(-10, 10, 500)
y_val = torch.cos(x_val.data)**2

plt.figure(figsize=(10,5))
plt.plot(x_val.numpy(), y_val.numpy(), 'o')
plt.title('Валидационная выборка')
plt.show()
```

```python
x_val.unsqueeze_(1)
y_val.unsqueeze_(1)
```

```python
class OurNet(torch.nn.Module):
    def __init__(self, n_hid_n):
        super(OurNet, self).__init__()
        self.fc1 = torch.nn.Linear(1, n_hid_n)
        self.act1 = torch.nn.Sigmoid()
        self.fc2 = torch.nn.Linear(n_hid_n, n_hid_n)
        self.act2 = torch.nn.Sigmoid()
        self.fc3 = torch.nn.Linear(n_hid_n,1)
        #1 вход -> n-нейронов -> n-нейронов -> n-нейронов-> 1 выход 
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        x = self.fc3(x)
        return x
```

```python
our_net = OurNet(16)
```

```python
def predict(net, x, y):
    y_predict = net.forward(x)
    
    plt.figure(figsize = (10,5))
    plt.plot(x.numpy(), y.numpy(), 'o', c = 'g', label = 'То что должно быть')
    plt.plot(x.numpy(), y_predict.data.numpy(), 'o', c = 'r', label = 'Предсказания')
    plt.legend(loc = 'upper left')
    plt.show()
```

```python
predict(our_net, x_val, y_val)
```

```python
optimizer = torch.optim.Adam(our_net.parameters(), lr=0.001)
```

```python
def loss(pred, true):
    sq = (pred - true)**2
    return sq.mean()
```

```python
for e in range(50000):
    optimizer.zero_grad()
    
    y_pred = our_net.forward(x_train)
    loss_val = loss(y_pred, y_train)
    print(loss_val)
    
    loss_val.backward()
    optimizer.step()
```

```python
predict(our_net, x_val, y_val)
```
